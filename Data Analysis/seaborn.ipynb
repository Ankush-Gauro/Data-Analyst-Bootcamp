```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
import matplotlib.pyplot as plt
```

```python
tips
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatterplot of Total Bill vs Tip')
plt.show()
```

```python
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Lineplot of Size vs Total Bill')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Barplot of Day vs Total Bill')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Boxplot of Day vs Total Bill')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violinplot of Day vs Total Bill')
plt.show()
```

```python
sns.histplot(x='total_bill', data=tips, bins=20, kde=True)
plt.title('Histogram of Total Bill with KDE')
```

```python
sns.kdeplot(x='total_bill', data=tips, fill=True)
```

```python
sns.pairplot(tips)
```

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
